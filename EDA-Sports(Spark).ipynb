**Matches played at each venue**

```python
plt.figure(figsize=(14,8))
sns.countplot(x='venue', data=df1)
plt.title("No. of matches played at each venue :")
text="Observation=Eden Gardens hosted most matches"
fig=plt.figure
plt.figtext(0.5, -0.45,text, wrap=True, horizontalalignment="center",fontsize=12)
plt.xticks(rotation=90)

plt.show()
```

**Matches played in each season**

```python
plt.figure(figsize=(14,8))
sns.countplot(x='season', data=df1, palette='Blues')
plt.title("No. of matches played each season")
text="Observation-In Season 2013, max. matches were played "
fig=plt.figure
plt.figtext(0.5, -0.01,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show()
```

**Matches played by each team**

```python
plt.figure(figsize=(14,8))
sns.countplot(x='team1', data=df1)
plt.title("Matches played by each team")
text="Observation-Mumbai Indians played most matches"
fig=plt.figure
plt.figtext(0.5, -0.25,text, wrap=True, horizontalalignment="center",fontsize=12)
plt.xticks(rotation=90)

plt.show()
```

**Number of tosses won by each team**

```python
plt.figure(figsize=(14,8))
sns.countplot(x='toss_winner', data=df1)
plt.title("Tosses won by each team")
text="Observation-Mumbai Indians won most tosses"
fig=plt.figure
plt.figtext(0.5, -0.25,text, wrap=True, horizontalalignment="center",fontsize=12)
plt.xticks(rotation=90)

plt.show()
```

**Wins per team**

**Player of the matches**

```python
plt.subplots(figsize=(14,8))
player = df1.player_of_match.value_counts()[:15]
sns.barplot(x=player.index, y=player, orient='v', palette=
           'Reds')
plt.title("Top Player of matches")
text="Observation-Chris Gayle has most Player of Matches award"
plt.figtext(0.5, -0.05,text, wrap=True, horizontalalignment="center",fontsize=12)
plt.xticks(rotation=90)

plt.show()
```

**Toss Decision**

```python
plt.subplots(figsize=(12,6))
toss = df1.toss_decision.value_counts()
label = (np.array(toss.index))
size = (np.array((toss/toss.sum())*100))
colors=['orange','yellow']
plt.pie(size, labels=label,colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Toss Decision %")
text="Observation-Batting first was chosen in most matches"
plt.figtext(0.5, 0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show()
```

**Win % if field first**

```python
plt.subplots(figsize=(12,6))
wins = (df1.win_by_wickets>0).sum()
loss = (df1.win_by_wickets==0).sum()
label = ['wins','loss']
size = [(wins/(wins+loss))*100, (loss/(wins+loss))*100]
colors=['lightgreen','red']
plt.pie(size, labels=label, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title("Win% fielding first")
text="Observation-53.7% of matches were won who chose to field first"
plt.figtext(0.5, 0.1,text, wrap=True, horizontalalignment="center",fontsize=12)


plt.show()
```

**Top Umpires**

## Team Analysis ##

**Teams scoring most runs**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('batting_team')['total_runs'].agg('sum').reset_index().sort_values(by='total_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['batting_team'])
r = np.array(runs['total_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='LightBlue')
ax.set_xlabel("Batting Team")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Total Runs")
ax.set_title("Teams scoring top runs")
text="Mumbai Indians have Scored Maximum Runs yet"
plt.figtext(0.5, -0.2,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

## Batsman Analysis ##

**Top run scorers**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('batsman')['batsman_runs'].agg('sum').reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['batsman'])
r = np.array(runs['batsman_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='Orange')
ax.set_xlabel("Batsman")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Runs")
ax.set_title("Top run scorers")
text="Virat Kohli Scored Maximum Runs yet"
plt.figtext(0.5, -0.05,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show

```

**Players hitting most 4s**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('batsman')['batsman_runs'].agg(lambda x: (x==4).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['batsman'])
r = np.array(runs['batsman_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='Purple')
ax.set_xlabel("Batsman")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Runs")
ax.set_title("Top players hitting most 4s")
text="S Dhawan Scored Maximum 4s yet"
plt.figtext(0.5, -0.05,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Players hitting most 6s**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('batsman')['batsman_runs'].agg(lambda x: (x==6).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['batsman'])
r = np.array(runs['batsman_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='Blue')
ax.set_xlabel("Batsman")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Runs")
ax.set_title("Top players hitting most 6s")
text="CH Gayle Scored Maximum 6s yet"
plt.figtext(0.5, -0.05,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Players playing least dots**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('batsman')['batsman_runs'].agg(lambda x: (x!=0).sum()).reset_index().sort_values(by='batsman_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['batsman'])
r = np.array(runs['batsman_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='LightGreen')
ax.set_xlabel("Batsman")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Runs")
ax.set_title("Players playing least dots")
text="V Kohli played least dots yet"
plt.figtext(0.5, -0.05,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

## Bowlers Analysis ##

**Top Wicket takers**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('bowler')['player_dismissed'].count().reset_index().sort_values(by='player_dismissed',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['bowler'])
r = np.array(runs['player_dismissed'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='forestgreen')
ax.set_xlabel("Bowler")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Wickets")
ax.set_title("Players scoring least dots")
text="SL Malinga has taken most wickets yet"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Bowlers bowling most dots**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('bowler')['total_runs'].agg(lambda x: (x==0).sum()).reset_index().sort_values(by='total_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['bowler'])
r = np.array(runs['total_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='violet')
ax.set_xlabel("bowler")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Runs")
ax.set_title("Bowler delivering most dots")
text="Harbhajan Singh bowled most dots yet"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Bowlers bowling most balls**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('bowler')['ball'].count().reset_index().sort_values(by='ball',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['bowler'])
r = np.array(runs['ball'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='slategrey')
ax.set_xlabel("Bowler")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Balls")
ax.set_title("Bowlers bowling most balls")
text="Harbhajan Singh bowled most balls yet"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Bowlers bowling most extras**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('bowler')['extra_runs'].agg(lambda x: (x>0).sum()).reset_index().sort_values(by='extra_runs',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['bowler'])
r = np.array(runs['extra_runs'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='aqua')
ax.set_xlabel("Bowler")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("Extras")
ax.set_title("Bowlers bowling most extras")
text="SL Malings has most extras yet"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Bowlers dismissing batsmen by kind=bowled**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('bowler')['dismissal_kind'].agg(lambda x: (x=='bowled').sum()).reset_index().sort_values(by='dismissal_kind',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['bowler'])
r = np.array(runs['dismissal_kind'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='pink')
ax.set_xlabel("Bowler")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("count")
ax.set_title("Bowlers dismissing batsmen by kind=bowled")
text="SL Malinga dismissed most batsmen by kind=bowled"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

## Dismissal Kinds Analysis

**Dismissal Kind % of total dismissals**

```python
df2.dismissal_kind.unique()
```

```python
plt.subplots(figsize=(14,8))

caught = (df2.dismissal_kind == 'caught').sum()
bowled = (df2.dismissal_kind == 'bowled').sum()
runout = (df2.dismissal_kind == 'run out').sum()
lbw = (df2.dismissal_kind == 'lbw').sum()
cb = (df2.dismissal_kind == 'caught and bowled').sum()
stumped = (df2.dismissal_kind == 'stumped').sum()
r = (df2.dismissal_kind == 'retired hurt').sum()
hit = (df2.dismissal_kind == 'hit wicket').sum()
obs = (df2.dismissal_kind == 'obstructing the field').sum()

label = ['bowled', 'obstructing field', 'caught', 'hit wicket', 'runout','retired hurt','lbw','caught & bowled','stumped']
total = (df2.dismissal_kind).count()
size = [(bowled/total)*100, (obs/total)*100, (caught/total)*100, (hit/total)*100, (runout/total)*100, (r/total)*100, (lbw/total)*100, (cb/total)*100, (stumped/total)*100]
colors=['lime','blue','darkturquoise','orange','pink','yellow','grey','ivory','violet']
plt.pie(size, labels=label, colors=colors, autopct='%1.1f%%', startangle=90, rotatelabels=90)
plt.title("Dimissal Kind% out of total dismissals", y=1.2)

text="Observation-60.5% of wickets were taken by catch"
plt.figtext(0.5, 0.05,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show()
```

**Fielders who caught most balls**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('fielder')['dismissal_kind'].agg(lambda x: (x=='caught').sum()).reset_index().sort_values(by='dismissal_kind',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['fielder'])
r = np.array(runs['dismissal_kind'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='lime')
ax.set_xlabel("Fielder")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("count")
ax.set_title("Top fielders who caught most balls")
text="KD Karthik caught most of the balls"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```

**Top Wicket Keepers**

```python
fig, ax = plt.subplots(figsize=(14,8))
runs = df2.groupby('fielder')['dismissal_kind'].agg(lambda x: (x=='stumped').sum()).reset_index().sort_values(by='dismissal_kind',ascending=False).reset_index(drop=True)
runs = runs.iloc[:10]
label = np.array(runs['fielder'])
r = np.array(runs['dismissal_kind'])
i = np.arange(len(label))
w = 0.9
bar = ax.bar(i, r, width=w, color='brown')
ax.set_xlabel("Wicket Keeper")
ax.set_xticks(i+(w/2.))
ax.set_xticklabels(label, rotation='vertical')
ax.set_ylabel("count")
ax.set_title("Top Wicket Keepers")
text="MS Dhoni has stumped maximum times"
plt.figtext(0.5, -0.1,text, wrap=True, horizontalalignment="center",fontsize=12)

plt.show
```
